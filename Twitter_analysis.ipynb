```python
#naming the columns and reading the dataset again
column_names = ['target','id','date','flag','user','text']
twitter_data = pd.read_csv('/content/training.1600000.processed.noemoticon.csv',names=column_names,encoding='ISO-8859-1')
twitter_data.head()
```

Convert the target "4" to "1"

```python
twitter_data.replace({'target':{4:1}},inplace=True)
twitter_data.head(10)
```

```python
twitter_data['target'].value_counts()
```

**Stemming**

Stemming is the process of reducing a word to its root word

eg:actor,actress,acting ->act

0->neg tweet 1->positive tweet

```python
port_stem = PorterStemmer()
```

```python
def stemming(content):
  stemmed_content = re.sub('[^a-zA-Z]',' ',content) #regular expression. remove everything which is not from a-z(upper and lower)
  stemmed_content = stemmed_content.lower()
  stemmed_content = stemmed_content.split()
  stemmed_content = [port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
  stemmed_content = ' '.join(stemmed_content)
  return stemmed_content



```

```python
twitter_data['stemmed_content'] = twitter_data['text'].apply(stemming)
```

```python
twitter_data.head()
```

```python
print(twitter_data['stemmed_content'])
```

```python
print(twitter_data['target'])
```

```python
# separating the data and label
X = twitter_data['stemmed_content']
Y = twitter_data['target'].values
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data to training data and test data

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size = 0.2,stratify=Y,random_state = 2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

```python
print(X_train)
```

```python
print(X_test)
```

```python
#converting textual data to numerical data
vectorizer = TfidfVectorizer()
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)

```

```python
print(X_train)
```

```python
print(X_test)
```

Training the ml model - Logisitic Regression

```python
model = LogisticRegression(max_iter = 1000)
model.fit(X_train,Y_train)
```

```python
#Accuracy score on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(Y_train,X_train_prediction)
```

```python
print("Accuracy score on the training data:",training_data_accuracy)
```

```python
#accuracy score on the test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(Y_test,X_test_prediction)
print("Accuracy score on the test data:",test_data_accuracy)
```

Model Accuracy = 77.8%.

Saving the trained model

```python
import pickle
filename = 'trained_model.sav'
pickle.dump(model,open(filename,'wb'))
```
